```python
def createWaveplot(data,sr,e):
    mplt.figure(figsize=(10,3))
    mplt.title('Waveplot for audio with {} emotion'.format(e),size=15)
    librosa.display.waveshow(data,sr=sr,color='b')
    mplt.show()
```

```python
def createSpectrogram(data,sr,e):
    X=librosa.stft(data)
    Xdb=librosa.amplitude_to_db(abs(X))
    mplt.figure(figsize=(12,3))
    mplt.title('Spectrogram for audio with {} emotion'.format(e),size=15)
    librosa.display.specshow(Xdb,sr=sr,x_axis='time',y_axis='hz')
    mplt.colorbar()
```

```python
import librosa
import numpy as np

def extractFeature(fileName, mfcc=True, chroma=True, mel=True):
    X, sampleRate = librosa.load(fileName, res_type='kaiser_fast')
    result = np.array([])

    if chroma:
        stft = np.abs(librosa.stft(X))
        chroma = np.mean(librosa.feature.chroma_stft(S=stft, sr=sampleRate).T, axis=0)
        result = np.hstack((result, chroma))

    if mfcc:
        mfccs = np.mean(librosa.feature.mfcc(y=X, sr=sampleRate, n_mfcc=40).T, axis=0)
        result = np.hstack((result, mfccs))

    if mel:
        mel = np.mean(librosa.feature.melspectrogram(y=X, sr=sampleRate).T, axis=0)
        result = np.hstack((result, mel))

    return result
```

```python
import glob
import os
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import numpy as np

emotion = {
    '01': 'neutral',
    '02': 'calm',
    '03': 'happy',
    '04': 'sad',
    '05': 'angry',
    '06': 'fearful',
    '07': 'disgust',
    '08': 'surprised'
}

def loadData(test_size=0.23, random_state=9):
    observedEmotions = ['calm', 'happy', 'fearful', 'disgust']
    x, y = [], []

    for file in glob.glob(r"C:\Users\User\OneDrive\Desktop\python\archive\DATA\Actor_*\*.wav"):
        fileName = os.path.basename(file)
        emotion_code = fileName.split("-")[2]
        emotion_label = emotion[emotion_code]

        if emotion_label not in observedEmotions:
            continue

        feature = extractFeature(file, mfcc=True, chroma=True, mel=True)
        x.append(feature)
        y.append(emotion_label)

    # Use LabelEncoder to convert string labels to numerical format
    label_encoder = LabelEncoder()
    y_encoded = label_encoder.fit_transform(y)

    x_train, x_test, y_train, y_test = train_test_split(
        np.array(x), y_encoded, test_size=test_size, random_state=random_state
    )

    return x_train, x_test, y_train, y_test
```

```python
#splitting dataset
xTrain,xTest,yTrain,yTest=loadData(test_size = 0.23)
print(xTrain.shape)
print(len(yTrain))
```

```python
#getting the shape of the training and testing datasets
print((xTrain.shape[0], xTest.shape[0]))
```

```python
#Getting the number of features extracted
print(f'Features extracted: {xTrain.shape[1]}')
```

```python
#Intialize the MultiLayer-Perception Classifer
model=MLPClassifier(alpha=0.01, batch_size=256, epsilon=1e-08,hidden_layer_sizes=(300,), learning_rate='adaptive',max_iter=500)
#model=MLPClassifer()
```

```python
from sklearn.preprocessing import LabelEncoder

# Convert yTrain to a NumPy array
yTrain = np.array(yTrain)

# Use LabelEncoder to convert string labels to numerical format
label_encoder = LabelEncoder()
yTrain = label_encoder.fit_transform(yTrain)

# Training the model
model.fit(xTrain, yTrain)
```

```python
#predicting for the test set
expected_Of_y=yTest
yPred=model.predict(xTest)
```

```python
print(metrics.confusion_matrix(expected_Of_y,yPred))
```

```python
print(classification_report(yTest, yPred))
```

```python
#Calculating the accuracy of our model
accuracy=accuracy_score(y_true=yTest, y_pred=yPred)
#printing the accuracy
print("Accuracy: {:.2f}%".format(accuracy*100))
```
